```python
from pycaret.datasets import get_data
dataSets = get_data('diabetes')
```

```python
from pycaret.clustering import *
ClusteringParameters = setup(dataSets)
```

```python
models()
```

```python
print("K-Means Clustering")
KMeanClusteringModel = create_model('kmeans')
print("K-Modes Clustering")
KModesClusteringModel = create_model('kmodes')
print("Affinity Propagation")
AffPropagationModel = create_model('ap')
print("Mean Shift Clustering")
MeanShiftModel = create_model('meanshift')
print("Spectral Clustering")
SpectralClusteringModel = create_model('sc')
print("Agglomerative Clustering")
AggloClusteringModel = create_model('hclust')
print("Density-Based Spatial Clustering")
DBScanClusteringModel = create_model('dbscan')
print("OPTICS Clustering	")
OpticsClusteringModel = create_model('optics')
print("Birch Clustering")
BirchClusteringModel = create_model('birch')
```

```python
MeanShiftPrediction = assign_model(MeanShiftModel)
MeanShiftPrediction
```

```python
plot_model(MeanShiftModel, plot='cluster')
```

```python
plot_model(MeanShiftModel, plot = 'tsne')
```

```python
plot_model(MeanShiftModel, plot = 'distribution')
```
